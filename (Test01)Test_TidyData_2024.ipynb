## 1. หาความแตกต่างของค่าเฉลี่ย Chlo_a ระหว่างปี 2018 และ 2019 สำหรับแต่ละสถานี (station)

```python
import pandas as pd

# อ่านไฟล์ Excel
df = pd.read_excel('test-TidyData_2024.xlsx')

# แสดงข้อมูล
print(df.head())
```

```python
# แยกปีจากคอลัมน์ 'month'
df['Year'] = df['month'].apply(lambda x: int(x.split('_')[1]))

# กรองข้อมูลเฉพาะปี 2018 และ 2019
df_2018 = df[df['Year'] == 2018]
df_2019 = df[df['Year'] == 2019]

# คำนวณค่าเฉลี่ย Chlo_a ของแต่ละ station ในแต่ละปี
mean_2018 = df_2018.groupby('station')['Chlo_a'].mean()
mean_2019 = df_2019.groupby('station')['Chlo_a'].mean()

# หาความแตกต่างของค่าเฉลี่ยระหว่างปี 2018 และ 2019
diff_mean = mean_2019 - mean_2018

# สร้าง DataFrame จากผลลัพธ์
result_df = pd.DataFrame({
    'Station': diff_mean.index,
    'Chlo_a_Mean_2018': mean_2018.values,
    'Chlo_a_Mean_2019': mean_2019.values,
    'Difference': diff_mean.values
})

# แสดงผล
for _, row in result_df.iterrows():
    print(f"Station {row['Station']}:")
    print(f"  Chlo_a in 2018: {row['Chlo_a_Mean_2018']:.4f}")
    print(f"  Chlo_a in 2019: {row['Chlo_a_Mean_2019']:.4f}")
    print(f"  Difference (2019 - 2018): {row['Difference']:.4f}")
    print()
```

## 2. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
import pandas as pd

# อ่านไฟล์ Excel
df = pd.read_excel('test-TidyData_2024.xlsx')

# แยกปีจากคอลัมน์ 'month'
df['Year'] = df['month'].apply(lambda x: int(x.split('_')[1]))

# คำนวณค่าเฉลี่ยของ Chlo_b สำหรับแต่ละ station ของแต่ละเดือน
average_Chlo_b = df.groupby(['Year', 'station'])['Chlo_b'].mean().reset_index()

# แสดงผลลัพธ์
print(average_Chlo_b)
```

```python
# แยกเดือนและปีออกจากคอลัมน์ 'month'
df[['Month', 'Year']] = df['month'].str.split('_', expand=True)
df['Year'] = df['Year'].astype(int)

# คำนวณค่าเฉลี่ยของ Chlo_b สำหรับแต่ละ station ของแต่ละเดือนและปี
average_Chlo_b = df.groupby(['Month', 'station'])['Chlo_b'].mean().reset_index()

# แสดงผลลัพธ์
print(average_Chlo_b)
```

```python
from tabulate import tabulate

# คำนวณค่าเฉลี่ยของ Chlo_b สำหรับแต่ละ station ของแต่ละเดือน
average_Chlo_b = df.groupby(['Month', 'station'])['Chlo_b'].mean().reset_index()

# แสดงผลลัพธ์ในรูปแบบตารางที่มีเส้น
print(tabulate(average_Chlo_b, headers='keys', tablefmt='grid'))
```
