```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates

```
pd.plotting.parallel_coordinates(df, 'Name');
```

## à¸›à¸£à¸±à¸š parameter: color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#FF00E0', '#00FFE2', '#ECFF00')) #html color
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#75c6de', '#5268cb', '#f9ea82'))
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('r', 'y', 'b')) #à¸•à¸±à¸§à¸¢à¹ˆà¸­à¸‚à¸­à¸‡à¸ªà¸µ R G B W C M Y K
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0.8,0), (0,0.7,0), (0.6,0,1))) #rgb value (r,g,b) | r,g,b -> [0,1]
```

## à¸›à¸£à¸±à¸š parameter: colormap

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='cool')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='plasma')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

```
np_data = df.iloc[:,:-1].to_numpy() #à¹€à¸¥à¸·à¸­à¸à¹à¸›à¸¥à¸‡à¸—à¸¸à¸à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸¢à¸à¹€à¸§à¹‰à¸™à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸ªà¸¸à¸”à¸—à¹‰à¸²à¸¢ à¹€à¸žà¸£à¸²à¸°à¸¡à¸±à¸™à¹€à¸›à¹‡à¸™à¸Šà¸·à¹ˆà¸­
np_data
```

# Visualize array data à¸”à¹‰à¸§à¸¢ pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #à¸à¸³à¸«à¸™à¸”à¸‚à¸™à¸²à¸”à¸‚à¸­à¸‡ figure

plt.imshow(np_data[:50,:],interpolation='nearest')
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest')
```

```
plt.imshow(np_data[50:100,:], interpolation='nearest', cmap='plasma')
```
